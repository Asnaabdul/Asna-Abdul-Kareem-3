5. another star pattern n=3

```python
n=3
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")
    for k in range(i + 1):
        print("* ", end="")
    print()
```

2. celsius to farhenheit

```python
x=int(input("Enter temperature in Celsius: "))
print("celsius is:",x)
def convert(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit
print("fahrenheit is:",convert(x))
```

3.removing a word

```python
def remove_and_strip(word_list, target):
    stripped_list = [w.strip() for w in word_list]
    
    if target.strip() in stripped_list:
        stripped_list.remove(target.strip())
    
    return stripped_list

words = ["  apple ", "banana", "  cherry  ", "date "]
target_word = " banana "

result = remove_and_strip(words, target_word)
print("Updated list:", result)
```
